```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil import parser

k = pd.read_csv('k_league_fw_goals.csv', index_col=0)
l = pd.read_csv('liverpool_14_15.csv', index_col=0, date_parser=parser.parse)
c = pd.read_csv('seac_coffee.csv')
```

```python
plt.plot(l['CumWin'])
```

```python
plt.plot(l[['CumWin', 'CumLost']])
```

```python
plt.plot(c[['A', 'C', 'E', 'S']])
```

```python
plt.bar(np.arange(4), c[['A', 'C', 'E', 'S']].sum().values)
```

```python
data = c[['A', 'C', 'E', 'S']].sum().values
index = np.arange(len(data))
bar_width = 0.5
plt.bar(index, data, width=bar_width, color='r')
plt.title('# Stores by Coffee Shop')
plt.xlabel('Coffee Shop')
plt.ylabel('# Store')
plt.xticks(index + bar_width - 0.25, ('A', 'C', 'E', 'S'))
```

```python
bar_height = 0.5
plt.barh(index, data, height=bar_height, color='g')
plt.title('# Stores by Coffee Shop')
plt.ylabel('Coffee Shop')
plt.xlabel('# Store')
plt.yticks(index + bar_width - 0.25, ('A', 'C', 'E', 'S'))
```

```python
labels = ('Angelinus', 'Cafe Bene', 'Ediya', 'Starbucks')
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90, explode=explode)
plt.axis('equal')
```

```python
fig, ax = plt.subplots()
ax.scatter(k['play'], k['goal'], alpha=0.5)
ax.set_xlabel('play')
ax.set_ylabel('goal')
ax.set_title('Play vs Goal')
```

```python
with plt.xkcd():
    bar_height = 0.5
    plt.barh(index, data, height=bar_height, color='g')
    plt.title('# Stores by Coffee Shop')
    plt.ylabel('Coffee Shop')
    plt.xlabel('# Store')
    plt.yticks(index + bar_width - 0.25, ('A', 'C', 'E', 'S'))
```
